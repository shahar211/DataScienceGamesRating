```python
from autosklearn.classification import AutoSklearnClassifier
```

```python
df= pd.read_csv('autotest.csv', index_col = 0)
```

```python
df.drop(["Name"], axis = 1, inplace = True)
```

```python
df
```

```python
df.describe()
```

```python
X = df.drop(["Rating"],axis=1)
y = df["Rating"]
```

```python
# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=33)
```

```python
print(f'X Training Shape:',X_train.shape)
print(f'y Training Shape:',y_train.shape)
print('***************************')
print(f'X Testing Shape y:',X_test.shape)
print(f'y Testing Shape y:',y_test.shape)
```

```python
import sklearn.metrics

import autosklearn.regression
```

```python
clf = autosklearn.regression.AutoSklearnRegressor(time_left_for_this_task=420, 
                              per_run_time_limit=9, 
                              ensemble_size=1, 
                              initial_configurations_via_metalearning=0)
clf.fit(X_train, y_train)
```

```python
clf.score(X_train, y_train)
```

```python
test_predictions = clf.predict(X_test)
print("Test R2 score:", sklearn.metrics.r2_score(y_test, test_predictions))
```
